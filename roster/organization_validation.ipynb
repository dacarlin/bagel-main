```python
import pandas 

allowed = pandas.read_csv( 'allowed_mutants.txt', header=0, index_col=0 )
current_roster = pandas.read_csv( 'current_roster.csv' )
```

```python
current_roster.sample( 3 )
```

```python
def is_good( label ):
    if label in allowed.index:
        return 1
    else:
        return 0
```

```python
print 't_name\tallowed\t'

for idx, series in current_roster.iterrows():
    tube_name = series.tube_name
    
    print tube_name, '\t', 
    print is_good( tube_name ), '\t', 
    print ''
    
```

```python
corrected = pandas.read_csv( 'corrected_roster.csv' )

print corrected.columns

print 'correct\tallowed\t'

for idx, series in corrected.iterrows():
    
    if is_good( series.correct_name ):
        pass 
    
    else:
        print series.correct_name, '\t', 
        print is_good( series.correct_name ), '\t', 
        print ''
    
```
